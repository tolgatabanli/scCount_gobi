```python
import scanpy as sc

# 1. Load the data (assumes files are in the current directory '.')
# Scanpy looks for 'matrix.mtx', 'barcodes.tsv', and 'genes.tsv' or 'features.tsv'
adata = sc.read_10x_mtx('.', var_names='gene_symbols', cache=True)

# 2. Basic Preprocessing
sc.pp.filter_cells(adata, min_genes=200)
sc.pp.filter_genes(adata, min_cells=3)
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)

# 3. Dimensionality Reduction
sc.pp.highly_variable_genes(adata, n_top_genes=2000)
adata = adata[:, adata.var.highly_variable]
sc.pp.scale(adata, max_value=10)
sc.tl.pca(adata, svd_solver='arpack')

# 4. Neighbors and UMAP
sc.pp.neighbors(adata, n_neighbors=10, n_pcs=40)
sc.tl.umap(adata)

# 5. Plot
sc.pl.umap(adata, color=None, title="UMAP of Loaded MTX Data")
```
